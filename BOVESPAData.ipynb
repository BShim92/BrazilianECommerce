```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import requests
from pprint import pprint
```

```python
url = 'https://www.quandl.com/api/v3/datasets/BCB/7.json?api_key=2vRuu5GNchksqBdcAHzc'
response = requests.get(url).json()
```

```python
ogdf = pd.DataFrame(response['dataset']['data'])
```

```python
columns = {
    0: 'date',
    1: 'value'}
ogdf = ogdf.rename(columns = columns)
ogdf.head()

# dfchange = pd.DataFrame(columns = [['date', 'change']])
# dfchange
```

```python
ogdf['value'] = pd.to_numeric(ogdf['value'])
ogdf = ogdf.sort_values(by = 'date', ascending = True)
ogdf['difference'] = ogdf['value'].diff(periods = 5)
ogdf['movingaverage'] = ogdf['value'].rolling(window=5).mean()
ogdf
```
